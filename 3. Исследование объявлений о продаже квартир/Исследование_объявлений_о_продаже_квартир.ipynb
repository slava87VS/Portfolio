## Замена типа данных

**Произведем замену типов данных согласно выводам в первой части**

```python
data['last_price'] = pd.to_numeric(data['last_price'], downcast="integer")
data['total_area'] = pd.to_numeric(data['total_area'], downcast="integer")
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%d %H:%M:%S')
data['floors_total'] = pd.to_numeric(data['floors_total'], downcast="integer")
data['living_area'] = pd.to_numeric(data['living_area'], downcast="integer")
data['is_apartment'] = data['is_apartment'].astype(bool)
data['kitchen_area'] = pd.to_numeric(data['kitchen_area'], downcast="integer")
data['balcony'] = pd.to_numeric(data['balcony'], downcast="integer")
data['airports_nearest'] = pd.to_numeric(data['airports_nearest'], downcast="integer")
data['cityCenters_nearest'] = pd.to_numeric(data['cityCenters_nearest'], downcast="integer")
data['parks_around3000'] = pd.to_numeric(data['parks_around3000'], downcast="integer")
data['parks_nearest'] = pd.to_numeric(data['parks_nearest'], downcast="integer")
data['ponds_around3000'] = pd.to_numeric(data['ponds_around3000'], downcast="integer")
data['ponds_nearest'] = pd.to_numeric(data['ponds_nearest'], downcast="integer")
data['days_exposition'] = pd.to_numeric(data['days_exposition'], downcast="integer")
```

## Предобработка данных

**Столбец ceiling_height** пропущенные значения меняем на медиану. Для расчета медианы сразу переведем тип данных из object в float. Затем выведем первые 10 строк

```python
data['ceiling_height'].isnull().sum()
```

```python
data['ceiling_height'] = data['ceiling_height'].astype(float)
```

```python
data.pivot_table(index='rooms', columns='floors_total', values='ceiling_height', aggfunc='median')
```

```python
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby(['rooms','floors_total'])['ceiling_height'].transform('median'))
data.head(10)
```

```python
data['ceiling_height'].isnull().sum()
```

```python
data['ceiling_height'] = data['ceiling_height'].astype(float)
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())
data['ceiling_height'].isnull().sum()
```

- **balcony**
**Если человек не указал число балконов — скорее всего их нет, такие пропуски правильно заменить на 0.**

```python
data['balcony'] = data['balcony'].fillna(0)
```

- **is_apartment**
**пропущенные значения заменим на False, большинство недвижимости это обычные квартиры, вероятно продавцы по умолчанию не пишут апартаменты это или нет**

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
```

**Найдем жилую площадь**

Предлагаю пропущенные значения в таком случае заменить следующим образом: вычислить средний процент, который занимает жилая площадь в общей площади. Для этого сделаем срез таблица где пристутсвуют значения:


```python
living_area_table = data.query('living_area !="Nan"')
part = living_area_table['living_area'].sum() / living_area_table['total_area'].sum()
data['living_area'] = data['living_area'].fillna(data['total_area']*part)
data
```

- **kitchen_area**

**Найдем площадь кухни**

С площадью кухни поступим примерно также, только рассчитаем процент площади кухни от НЕЖИЛОЙ площади. Но сперва посмотрим все возможные корреляции:

```python
table_kitchen_area = data.query('kitchen_area != "Nan"').copy()
table_kitchen_area['not_living_area'] = table_kitchen_area['total_area'] - table_kitchen_area['living_area']
table_kitchen_area['kitchen_area'].corr(table_kitchen_area['not_living_area']) #считаем корреляцию площади кухни и нежилой площади
```

```python
table_kitchen_area['kitchen_area'].corr(table_kitchen_area['total_area']) #считаем корреляцию площади кухни и общей площади
```

```python
table_kitchen_area['kitchen_area'].corr(table_kitchen_area['living_area']) #считаем корреляцию площади кухни и жилой площади
```

**Да, гипотеза подтвердилась, будем использовать процент от нежилой площади:**

```python
part_kitchen = table_kitchen_area['kitchen_area'].sum() / table_kitchen_area['not_living_area'].sum()
data['kitchen_area'] = data['kitchen_area'].fillna(data['total_area']*part_kitchen)
data
```
