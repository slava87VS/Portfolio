### Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на обучающей выборке для признака rougher.output.recovery (эффективность обогащения чернового концентрата). Найдем MAE между нашими расчётами и значением признака. Опишем выводы.

- **C — доля золота в концентрате после флотации/очистки;**
- **F — доля золота в сырье/концентрате до флотации/очистки;**
- **T — доля золота в отвальных хвостах после флотации/очистки.**

```python
c = train['rougher.output.concentrate_au']
f = train['rougher.input.feed_au']
t = train['rougher.output.tail_au']

recovery_calculated = ((c * (f-t)) / (f * (c-t))) * 100
recovery_column = train['rougher.output.recovery']
print('MAE', mean_absolute_error(recovery_column, recovery_calculated))
```

***Вывод: MAE достаточно низкое, значит отклонение между расчетным значением rougher.output.recovery и этим же значением в таблице практически совпадает***

### Проанализируем признаки, недоступные в тестовой выборке. Выясним что это за параметры. К какому типу относятся

```python
def for_merge(data):
    s = data.drop('date', axis=1)
    s = s.T
    s = pd.DataFrame(s)
    s = s.rename_axis('parametr').reset_index()
    s['mean'] = s.iloc[:, 1:-1].mean(axis=1)
    s = s.drop(s.iloc[:, 1:-1], axis=1)
    return s
```

```python
test_for_merge = for_merge(test).rename(columns={'mean' : 'test'})
```

```python
train_for_merge = for_merge(train).rename(columns={'mean' : 'train'})
```

```python
full_for_merge = for_merge(full).rename(columns={'mean' : 'full'})
```

```python
all = full_for_merge.merge(test_for_merge, on='parametr', how='outer')
all = all.merge(train_for_merge, on='parametr', how='outer')
all.head(50)
```

### Напишем функцию для вычисления итоговой sMAPE

```python
def smape(target, predict):
    error = np.mean(abs(target - predict)/((abs(target)+abs(predict))/2))*100
    return error
def smape_final(rougher,final):
    s = (0.25*(rougher) + 0.75*final)
    return s
```
